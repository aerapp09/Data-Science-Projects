```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('countries.csv')
data.head()
```

```python
data_2007 = data[data.year == 2007]
data_2007.head()
```

```python
top10 = data_2007.sort_values('population', ascending=False).head(10)
top10
```

```python
top10_GDP = (top10.gdpPerCapita * top10.population) / 10**9
```

```python
plt.subplot(2,1,1)
x = range(10)
plt.bar(x, top10.population / 10**6)
plt.xticks([], [])
plt.title('GDP of 10 Most Populous Countries')
plt.legend(['Population in Millions'])

plt.subplot(2,1,2)
plt.bar(x, top10_GDP)
#plt.title('GDP of 10 Most Populous Countries')
plt.xlabel('Country')
plt.ylabel('GDP (Billions $)')
plt.xticks(x, top10.country, rotation='vertical')
plt.show()
```

```python
x = np.arange(10)

fig, ax1 = plt.subplots()

width = .3
plt.xticks(x, top10.country, rotation = 'vertical')
population = ax1.bar(x, top10.population / 10**6, width)
plt.ylabel('Population')

ax2 = ax1.twinx()
gdp= ax2.bar(x + width, top10.gdpPerCapita * top10.population/10**9, width, color='orange')
plt.ylabel('GDP')
plt.legend([population, gdp], 
          ['Population in Millions', 'GDP in Billions'])
figure = plt.gcf() 
plt.show()
```
