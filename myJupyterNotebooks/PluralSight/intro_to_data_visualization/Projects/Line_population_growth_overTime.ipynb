```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('countries.csv')
data.head()
```

```python
us = data[data.country == 'United States']
china = data[data.country == 'China']
```

```python
plt.plot(us.year, us.population/10**6 )
plt.plot(china.year, china.population/10**6)
plt.title("Absolute population Growth in United States and China")
plt.legend(['United States', 'China'])
plt.xlabel('year')
plt.ylabel('Population Growth')
```

```python
us.population.iloc[0]
```

```python
us.population/ us.population.iloc[0]
```

```python
us_growth = us.population / us.population.iloc[0] * 100
china_growth = china.population / china.population.iloc[0] *100
```

```python
plt.plot(us.year, us_growth )
plt.plot(china.year, china_growth)
plt.title("Relative population Growth in United States and China(first population = 100)")
plt.legend(['United States', 'China'])
plt.xlabel('year')
plt.ylabel('Population Growth')
```
