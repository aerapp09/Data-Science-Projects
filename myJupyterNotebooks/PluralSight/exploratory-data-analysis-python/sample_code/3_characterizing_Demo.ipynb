```python
np.random.seed(0)
```

```python
mu, sigma=0,1
```

```python
data = np.random.normal(mu, sigma, 1000)
```

```python
p = sns.kdeplot(data,
                color='#F15B2A',
                fill=True)
```

```python
p = sns.displot(data,
                color='#F15B2A',
                kde=True,
                fill=True)
```

```python
plot = sns.kdeplot(data,
                  color='#F15B2A',
                  fill=True)
```

```python
p = sns.boxplot(data,
               palette=['#F15B2A'])
```

```python
p = sns.boxplot(data,
               palette=["#F15B2A"])

for e in p.artists:
  r, g, b, a = e.get_facecolor()
  e.set_facecolor((r, g, b, .3))
  
for i, artist in enumerate(p.artists):
    artist.set_edgecolor("#F15B2A")
    
for j in range(0,6):
    line = p.lines[j]
    line.set_color("#F15B2A")
    line.set_mfc("#F15B2A")
    line.set_mec("#F15B2A")
```

```python
p = sns.violinplot(data,
                  palette=['#F15B2A'])
```
