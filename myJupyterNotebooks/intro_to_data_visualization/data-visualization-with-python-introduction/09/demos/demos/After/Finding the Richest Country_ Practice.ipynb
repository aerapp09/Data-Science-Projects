```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('countries.csv')
```

```python
data.head()
```

```python
data_country = data.groupby(['country']).mean()
data_asc= data_country.sort_values('gdpPerCapita', ascending=False)
data_top = data_asc.head(10)
data_top
```

```python
x = range(10)
plt.bar(x, data_top.gdpPerCapita)
plt.xticks(x, data_top.index, rotation='vertical')
plt.title('10 Most Rich Countries by GDP per Capita')
plt.ylabel('GDP per Capita')
plt.show()
```

```python
kuwait = data[data.country == 'Kuwait']
```

```python
kuwait.head()
```

```python
plt.plot(kuwait.year, kuwait.gdpPerCapita)
plt.title('Kuwait GDP per Capita')
plt.ylabel('GDP per Capita ($)')
plt.xlabel('year')
```

```python
plt.subplot(311)
plt.title('GDP per Capita')
plt.plot(kuwait.year, kuwait.gdpPerCapita)

plt.subplot(312)
plt.title('GDP in Billions')
plt.plot(kuwait.year, kuwait.population*kuwait.gdpPerCapita)

plt.subplot(313)
plt.title('Population')
plt.plot(kuwait.year, kuwait.population/10**6)
plt.tight_layout()
```

```python
kuwait_gdp = kuwait.population*kuwait.gdpPerCapita
plt.plot(kuwait.year, kuwait.population/kuwait.population.iloc[0] *100)
## this section at the end sets the first year as 100 as a basis for comparison
plt.plot(kuwait.year, kuwait_gdp/kuwait_gdp.iloc[0]*100)
plt.plot(kuwait.year, kuwait.gdpPerCapita / kuwait.gdpPerCapita.iloc[0]*100)
plt.title('GDP and Population Growth in Kuwait (first year = 100)')
plt.legend(['Population', 'GDP', 'gdpPerCapita'])
plt.show()
```
